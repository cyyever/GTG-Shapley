## run time pie chart

```python
dir_name='results/'
file_name='MR_OurMNIST_mlp_10_1.txt'
f=open(dir_name+file_name,'r')
lines=f.readlines()

index_s=0
while 'Total Run Time' not in lines[index_s]:
    index_s+=1

line=lines[index_s].split(':')[1]
total_t=float(line)

dura_tr=float(lines[index_s+1].split('=')[1])
dura_as=float(lines[index_s+3].split('=')[1])
dura_ev=float(lines[index_s+5].split('=')[1])
```

## utility in permutation

```python
dir_name='Simulation_temp/'
file_name='og_ut_dict.pkl'
f=open(dir_name+file_name,'rb')
U_all=pickle.load(f)

U=U_all['2']

N_perm=500
idxs=[i for i in range(1,11)]

marginal_u={i:[] for i in range(1,11)}

for t in range(N_perm):
    v=[0 for i in range(11)]
    v[0]=U[()]

    idxs_t=np.random.permutation(idxs)

    for j in range(1,11):
        C=idxs_t[:j]
        C=tuple(np.sort(C,kind='mergesort'))
        v[j]=U[C]

        marginal_u[j].append(v[j]-v[j-1])
```

## line chart

```python
fig,ax=plt.subplots()

labels=['Order %d'%i for i in range(11)]
x=[i for i in range(500)]

#log10
mu={}
for j in range(1,11):
    mu[j]=np.log10(np.array(marginal_u[j])+1.1e-5)

for j in range(1,11):
    ax.plot(x,mu[j],label=labels[j])
    
ax.set_xlabel('Permutatoins',size=10)
ax.set_ylabel('$log_{10}{Marginal Gain}$',size=10)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_title('Marginal Utility Gain of Participant in Order',fontweight='bold',fontsize=12)
    
    
plt.legend()


plt.savefig('Marginal Utility Gain of Participant in Order.eps',format='eps',bbox_inches='tight')


plt.show()

```
